```python
df.drop_duplicates(inplace=True)
```

```python
data = df.drop(["Name","Unnamed: 0","Establishment Year","Best Time to visit","Number of google review in lakhs","DSLR Allowed","Google review rating","Weekly Off","Type","City","State","time needed to visit in hrs","Entrance Fee in INR"],axis =1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Make sure the column is treated as string
data["Airport with 50km Radius"] = data["Airport with 50km Radius"].astype(str)

# Initialize and fit LabelEncoder
le_airport = LabelEncoder()
data["Airport with 50km Radius"] = le_airport.fit_transform(data["Airport with 50km Radius"])
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define categorical columns
categorical_cols = ["Zone", "Significance"]

# Clean the data
data[categorical_cols] = data[categorical_cols].fillna("Unknown").astype(str)

# Create the OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Build ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("onehot", encoder, categorical_cols)
    ],
    remainder="passthrough"
)

# Fit and transform
encoded_features = preprocessor.fit_transform(data)

# âœ… Get all output column names (including one-hot and passthrough)
feature_names = preprocessor.get_feature_names_out()

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_features, columns=feature_names)

# Preview the result
print(encoded_df.head())
```

```python
# Drop any columns like "Name" that shouldn't affect distance
X = encoded_df

from sklearn.neighbors import NearestNeighbors

# Train KNN
knn_model = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn_model.fit(X)
```

```python
import pandas as pd

# Define the same categorical columns used in your training
categorical_cols = ["Zone", "Significance"]

# Sample input from user
user_input = {
    "Zone": input("Enter Zone (e.g., Central, Northern, Southern): "),
    "Significance": input("Enter Significance (e.g., Historical, Cultural, Wildlife): "),
    "Airport with 50km Radius": float(input("Enter 1 if there is an airport within 50km, else 0: "))
}

# Convert input to DataFrame
user_df = pd.DataFrame([user_input])

# Ensure categorical columns are strings
user_df[categorical_cols] = user_df[categorical_cols].fillna("Unknown").astype(str)

# Apply the same encoder and preprocessor used earlier
user_encoded = preprocessor.transform(user_df)

# Convert to DataFrame (optional - only if you want to inspect)
user_encoded_df = pd.DataFrame(user_encoded, columns=preprocessor.get_feature_names_out())

# âœ… Now `user_encoded` (or `user_encoded_df`) is ready to be fed into your KNN model
# For example:
distances, indices = knn_model.kneighbors(user_encoded)

# Display results
print("\nðŸ” Top 5 Recommended Places:")
print(df.iloc[indices[0]][["Name", "City", "State","Type","Significance","Entrance Fee in INR","Google review rating","Zone"]])
```
